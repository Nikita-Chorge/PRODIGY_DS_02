```python
for i in range(len(original_data['date'])):
    date_str = str(original_data.iloc[i,1])
    date_str = date_str.replace('/','-')
    original_data.iloc[i,1] = date_str
```

```python
original_data.head()
```

```python
original_data['date']=pd.to_datetime(original_data['date'],format='%m-%d-%Y')
original_data.head()
```

```python
original_data.info()
```

```python
#seperately storing only month and year for analysis
original_data['Month_Year'] = original_data['date'].dt.strftime('%Y-%m')
```

```python
sns.histplot(data = original_data, x = 'item_name')
plt.xticks(rotation=90)
plt.xlabel('Item Name')
plt.title('Number of items sold')
plt.show()
```

```python
sns.boxplot(data = original_data, x = 'time_of_sale', y = 'transaction_amount')
plt.xlabel('Time of sale')
plt.ylabel('Transaction amount')
plt.show()
```

```python
sns.histplot(data = original_data, x = 'Month_Year')
plt.xticks(rotation=90)
plt.xlabel('Month Year')
plt.show()
```

```python
#counting the occurrences of unique values in the column
original_data['transaction_type'].value_counts()
```

```python
distribution = original_data['transaction_type'].value_counts(normalize=True)
original_data['transaction_type'].fillna(pd.Series(np.random.choice(distribution.index, p=distribution.values, size=len(original_data))), inplace=True)
```

```python
pd.pivot_table(original_data,values='transaction_amount',index='Month_Year',columns='item_name').plot.bar(y=original_data['item_name'].unique(),
figsize=(12,8),title='Revenue by Item Name')

plt.show()
```

```python
pd.pivot_table(original_data,values='quantity',index='item_type',columns='time_of_sale').round(2)
```

```python
pd.pivot_table(original_data,values='quantity',index='item_name',columns='time_of_sale').round(2)
```

```python
original_data.groupby('Month_Year')['transaction_amount'].sum().plot(kind='bar')
plt.title("Sales by Month")
plt.show()
```

```python
df = original_data.drop(['order_id'], axis =1)
sns.heatmap(df.corr())
```

```python
z=sns.kdeplot(original_data.groupby('Month_Year')['transaction_amount'].sum(), color='pink')
x = z.lines[0].get_xdata()
y = z.lines[0].get_ydata()
maxid = np.argmax(y)
plt.plot(x[maxid],y[maxid],'ro')
plt.show()
```

```python
palette = sns.color_palette("pastel", n_colors=2)
original_data.groupby('transaction_type')['transaction_amount'].sum().plot(kind='bar', figsize=(8,4), color=palette)
plt.show()
```

```python
palette = sns.color_palette("pastel", n_colors=5)
original_data.groupby('item_name')['quantity'].mean().plot(kind='bar', figsize=(8,4), color=palette)
plt.title("Avg daily Sales (item-wise)")
plt.show()
```
